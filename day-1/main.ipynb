```python
import numpy as np

point = 50
inputs = [
    "L68",
    "L30",
    "R48",
    "L5",
    "R60",
    "L55",
    "L1",
    "L99",
    "R14",
    "L82"
]

def read_file(input_path):
    with open(input_path) as f:
        data = f.readlines()
    return data

def rotations(inputs, start_point=50):
    point = start_point
    counter = 0
    for move in inputs:
        rot = int(move[1:])
        sp = point
        if move[0] == "L":
            point = point - rot
            if sp < rot and sp != 0:
                counter += abs(point // 100)
        else:
            point = point + rot
            counter += abs(point // 100)
        
        if point % 100 == 0:
            counter += 1
            print("zero point")
        
        point = point % 100
    print(counter)


rotations(inputs)
```

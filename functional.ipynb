## Python functional programming tools basic examples

All built in functions: https://docs.python.org/3/library/functions.html

### Map

```python
my_cars = ["volvo", "saab", "tesla", "lada", "vw"]

upper_cars = []

for car in my_cars:
    upper_cars.append(car.upper())

print(upper_cars)
```

Same but with map function

```python
upper_cars = list(map(str.upper, my_cars))
print(upper_cars)

capital_cars = list(map(str.capitalize, my_cars))
print(capital_cars)
```

### Example 2

```python
numbers = [1.234567, 1.234567, 1.234567, 1.234567, 1.234567]
rounded_numbers = list(map(round, numbers, range(1,6)))
print(rounded_numbers)
```

## Filter

### Example 1

```python
scores = [10, 15, 39, 67, 3, 43, 54]

def good_score(score):
    return score > 40

good_scores = list(filter(good_score, scores))
print(good_scores)
```

### Example 2

```python
finnish_words = ["tiski", "auto", "mallas", "oho", "tavat", "kioski"]

palindromes = list(filter(lambda word: word == word[::-1], finnish_words))
print(palindromes)
```

## Reduce

### Example 1

```python
from functools import reduce

numbers = [2, 5, 12, 4, 35, 16]

def sum_(first, second):
    return first + second

result = reduce(sum_, numbers)
print(result)

# Also possible to add initial values
result = reduce(sum_, numbers, 10)
print(result)
```

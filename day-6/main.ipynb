### Test inputs

```python
inputs = [
    '123 328  51 64 ',
    ' 45 64  387 23 ',
    '  6 98  215 314',
    '*   +   *   +' 
]
```

## Task 1

```python
from functools import reduce
def add(x, y):
    return x + y


inputs_grid = list(map(str.split, inputs))
symbols = inputs_grid[-1]
numbers = [[int(x) for x in row] for row in inputs_grid[:-1]]


sum_array = [int(sym == "*") for sym in symbols]
for row in numbers:
    for i, x in enumerate(row):
        if symbols[i] == "*":
            sum_array[i] *= x
        else:
            sum_array[i] += x



res = reduce(add, sum_array)
print(res)
```

```python
from functools import reduce

def add(x, y):
    return x + y

def mul(x, y):
    return x * y

# Numbers: [14 x 3] array
numbers = [[] for j in range(len(inputs[0]))]

for row_id, arr in enumerate(inputs[:-1]):
    for col_id, ch in enumerate(arr):
        if ch != ' ':
            numbers[col_id] += ch
    

cache = []
total = []
operation_id = 0
for i, arr in enumerate(numbers):
    joined_arr = "".join(arr)
    
    if len(joined_arr) > 0:
        cache.append(int(joined_arr))

    if len(joined_arr) == 0 or i == len(numbers)-1:
        # Do operation
        print(cache, symbols[operation_id])
        if symbols[operation_id] == "*":
            total.append(reduce(mul, cache))
        else:
            total.append(reduce(add, cache))
        operation_id += 1
        cache = []

print(reduce(add, total))
```

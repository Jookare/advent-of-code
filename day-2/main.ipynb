## Task 1

```python
inputs = [
    '11-22',
    '95-115',
    '998-1012',
    '1188511880-1188511890',
    '222220-222224',
    '1698522-1698528',
    '446443-446449',
    '38593856-38593862',
    '565653-565659',
    '824824821-824824827',
    '2121212118-2121212124'
]

invalid_ids = []
for r in inputs:
    start = r.split("-")[0]
    end = r.split("-")[1]
    
    # Remove odd numbers
    if len(start) % 2:
        start = "1"+len(start)*"0"
    if len(end) % 2:
        end = "9"*(len(end)-1)
    
    for value in range(int(start), int(end)):
        value = str(value)
        midpoint = len(str(value))//2
        flag = True
        for i in range(midpoint):
            if value[i] != value[midpoint+i]:
                flag = False
                break
        if flag:
            invalid_ids.append(value)

print(invalid_ids)
```

## Task 1

```python

invalid_ids = []
for r in inputs:
    start = r.split("-")[0]
    end = r.split("-")[1]
    
    # Remove odd numbers
    if len(start) % 2:
        start = "1"+len(start)*"0"
    if len(end) % 2:
        end = "9"*(len(end)-1)
    
    for value in range(int(start), int(end)+1):
        value = str(value)
        midpoint = len(str(value))//2
        flag = True
        for i in range(midpoint):
            if value[i] != value[midpoint+i]:
                flag = False
                break
        if flag:
            invalid_ids.append(value)

print(invalid_ids)
```

## Task 2

```python

invalid_ids = []
for r in inputs:
    start = r.split("-")[0]
    end = r.split("-")[1]
    
    for value in range(int(start), int(end)+1):
        value = str(value)
        l = len(value)
        
        c = l
        while c > 1:
            if l % c == 0:
                d = l//c
                if value == c * value[:d]:
                    invalid_ids.append(value)
                    break
            c -=1
        

print(invalid_ids)
```

# Functionals

## Task 1

```python
def check_invalid(value):
    # Find midpoint and slice value
    midpoint = len(value)//2
    first_half = value[:midpoint]
    second_half = value[midpoint:]

    # Check if all values are the same in both halves
    flag = all(a == b for a, b in zip(first_half, second_half))
    return flag

def normalize_range(start_s, end_s):
    # Remove odd numbers
    if len(start_s) % 2:
        start_s = "1"+len(start_s)*"0"
    if len(end_s) % 2:
        end_s = "9"*(len(end_s)-1)
    return int(start_s), int(end_s)
        
invalid_ids = []
for r in inputs:
    start_s, end_s = r.split("-")
    
    start, end = normalize_range(start_s, end_s)

    for value in range(start, end + 1):
        if check_invalid(str(value)):
            invalid_ids.append(int(value))

print(sum(invalid_ids))
```

## Task 2

```python
def split_factor(length):
    for l in range(1, length//2 + 1):
        if length % l == 0:
            yield l

def check_invalid_new(value):
    length = len(value)
    for l in split_factor(length):
        flag = value == (length // l) * value[:l]
        if flag:
            return True
    return False
 
invalid_ids = []
for r in inputs:
    start, end = map(int, r.split("-"))

    for value in range(start, end + 1):
        if check_invalid_new(str(value)):
            invalid_ids.append(int(value))


print(sum(invalid_ids))
```

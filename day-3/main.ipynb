```python
input = [
    '987654321111111',
    '811111111111119',
    '234234234234278',
    '818181911112111',
]


joltage = 0
for x in input:
    split = [int(dig) for dig in x]
    max_digit = max(split)
    position = split.index(max_digit)
    print(split)
    if position != len(x)-1:
        # If max digit is not last
        split = split[position+1:]
        second_digit = max(split)
        largest = int(str(max_digit) + str(second_digit))
    else:
        # Max digit is last
        split = split[:-1]
        first_digit = max(split)
        largest = int(str(first_digit) + str(max_digit))
    
    joltage += largest
        
print(joltage)

```

### Test inputs

```python
inputs = [
    '987654321111111',
    '811111111111119',
    '234234234234278',
    '818181911112111',
]
```

## Task 1

```python

# Quite naive solution
joltage = 0
for input in inputs:
    split = list(map(int, input))
    max_digit = max(split)
    position = split.index(max_digit)
    if position != len(input)-1:
        # If max digit is not last
        split = split[position+1:]
        second_digit = max(split)
        largest = int(str(max_digit) + str(second_digit))
    else:
        # Max digit is last
        split = split[:-1]
        first_digit = max(split)
        largest = int(str(first_digit) + str(max_digit))
    
    joltage += largest
        
print(joltage)

```

## Task 2

```python
def find_max_joltage(split, n):
    start_idx = 0
    index_list = []
    for i in range(1,n+1):
        end_idx = len(split)-(n-i)
        max_num = max(split[start_idx:end_idx])
        max_idx = split.index(max_num, start_idx, end_idx)
        index_list.append(max_idx)
        start_idx = max_idx+1
    
    values = [str(split[id]) for id in (index_list)]
    return int("".join(values))

joltage = 0
for input in inputs:
    split = list(map(int, input))
    joltage += find_max_joltage(split, 12)

print(joltage)
```
